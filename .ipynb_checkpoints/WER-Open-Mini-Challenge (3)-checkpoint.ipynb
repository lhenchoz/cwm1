In der unteren Tabelle sind die Variablen, welche für uns relevant sind.

| Variable          |                                  Beschreibung |
|:------------------|----------------------------------------------:|
| BETRIEBSTAG       | Datum, an welchem die Fahrt stattgefunden hat |
| LINIEN_ID         |                                   Fahrtnummer |

Wichtig ist, dass die Prognose Variablen abhängig von den Prognose-Status-Variablen sind. Diese können folgende Werte haben:

- UNBEKANNT (Keine Prognosen- und Ist-Zeiten für diesen und alle vorhergehenden Halte vorhanden)
- Leer (=PROGNOSE)
- PROGNOSE (Abfahrtsprognose)
- GESCHAETZT (berechnete Ist-Abfahrtszeit)
- REAL (effektive Ist-Abfahrtszeit)

Die Prognose Variablen sind also nur tatsächliche Ist-Zeiten, wenn der Prognose-Status REAL ist. Werte, welche bei uns UNBEKANNT sind, löschen wir, da wir nicht ermitteln können, ob der Zug ausgefallen ist oder ob die Übermittlung beim Halt nicht funktioniert hat (Open-Data-Plattform Mobilität Schweiz, o.J., Wie erkennt man, dass keine Echtzeitdaten übermittelt wurden?).

##### Umstrukturierung des Dataframes
Um ein Modell aus unseren Daten zu erstellen, müssen wir diese umstrukturieren. Wir selektieren für jede Fahrt von Liestal nach Brugg die jeweiligen Haltestellen mit dem gleichen BETRIEBSTAG und fassen so eine Fahrt kompakt in einer Zeile zusammen. Zusätzlich werden direkt die Zeitdifferenzen zwischen Soll- und Ist-Zeit berechnet, welche wir benötigen, um unsere Zufallsvariablen zu modellieren.

```python
#Datensatz mit allen Spalten leer erstellen
df_liestal = pd.DataFrame({
    'BETRIEBSTAG': datetime,
    'SOLL_AB_LST': datetime,
    'SOLL_AN_OLT': datetime,
    'IST_AN_OLT': datetime,
    'DELTA_AN_OLT': datetime,
    'SOLL_AB_OLT': datetime,
    'IST_AB_OLT': datetime,
    'DELTA_AB_OLT': datetime,
    'SOLL_AN_BRG': datetime,
    'IST_AN_BRG': datetime,
    'DELTA_AN_BRG': datetime}, index= [0])


count = 0

for i in range(0, len(df_liestal_raw)): # Durch df_liestal_raw iterieren
    # Überprüfen, ob ein Eintrag bereits mit dem jeweiligem BETRIEBSTAG existiert.
    summe = sum(df_liestal_raw.BETRIEBSTAG[i] == df_liestal.BETRIEBSTAG)
    if summe == 0:  # falls es noch keinen Eintrag gibt, einen neuen erstellen
        df_liestal.at[count, 'BETRIEBSTAG'] = df_liestal_raw.BETRIEBSTAG[i]


    # Hier werden die Daten aus dem Zug, je nach LINIEN_ID und HALTESTELLEN_NAME, einer Zelle im Dataframe zugewiesen
    if (df_liestal_raw.LINIEN_ID[i] == 959) & (df_liestal_raw.HALTESTELLEN_NAME[i] == "Liestal"):
        ind = df_liestal.index[df_liestal.BETRIEBSTAG == df_liestal_raw.BETRIEBSTAG[i]][0]
        df_liestal.at[ind, 'SOLL_AB_LST'] = df_liestal_raw.AB_PROGNOSE[i]

    if (df_liestal_raw.LINIEN_ID[i] == 959) & (df_liestal_raw.HALTESTELLEN_NAME[i] == "Olten"):
        ind = df_liestal.index[df_liestal.BETRIEBSTAG == df_liestal_raw.BETRIEBSTAG[i]][0]
        df_liestal.at[ind, 'SOLL_AN_OLT'] = df_liestal_raw.ANKUNFTSZEIT[i]
        df_liestal.at[ind, 'IST_AN_OLT'] = df_liestal_raw.AN_PROGNOSE[i]

    if (df_liestal_raw.LINIEN_ID[i] == 4761) & (df_liestal_raw.HALTESTELLEN_NAME[i] == "Olten"):
        ind = df_liestal.index[df_liestal.BETRIEBSTAG == df_liestal_raw.BETRIEBSTAG[i]][0]
        df_liestal.at[ind, 'SOLL_AB_OLT'] = df_liestal_raw.ABFAHRTSZEIT[i]
        df_liestal.at[ind, 'IST_AB_OLT'] = df_liestal_raw.AB_PROGNOSE[i]

    if (df_liestal_raw.LINIEN_ID[i] == 4761) & (df_liestal_raw.HALTESTELLEN_NAME[i] == "Brugg AG"):
        ind = df_liestal.index[df_liestal.BETRIEBSTAG == df_liestal_raw.BETRIEBSTAG[i]][0]
        df_liestal.at[ind, 'SOLL_AN_BRG'] = df_liestal_raw.ANKUNFTSZEIT[i]
        df_liestal.at[ind, 'IST_AN_BRG'] = df_liestal_raw.AN_PROGNOSE[i]

    # Hier müssen wir, um die NAs zu löschen, temporäre Werte einpflegen
    df_liestal.at[i, 'DELTA_AN_OLT'] = "1"
    df_liestal.at[i, 'DELTA_AB_OLT'] = "1"
    df_liestal.at[i, 'DELTA_AN_BRG'] = "1"

    if summe == 0:
        count += 1

# Hier entfernen wir die Reihen mit fehlenden Werten, da wir wie bereits erwähnt diese nicht als Ausfälle zählen können.
df_liestal = df_liestal.dropna()

# Hier werden die Zeitunterschiede zwischen Soll- und Ist-Zeiten berechnet.
df_liestal["DELTA_AB_OLT"] = (df_liestal["IST_AB_OLT"] - df_liestal["SOLL_AB_OLT"])
df_liestal["DELTA_AB_OLT"] = df_liestal["DELTA_AB_OLT"].dt.total_seconds()

df_liestal["DELTA_AN_OLT"] = (df_liestal["IST_AN_OLT"] - df_liestal["SOLL_AN_OLT"])
df_liestal["DELTA_AN_OLT"] = df_liestal["DELTA_AN_OLT"].dt.total_seconds()

df_liestal["DELTA_AN_BRG"] = (df_liestal["IST_AN_BRG"] - df_liestal["SOLL_AN_BRG"])
df_liestal["DELTA_AN_BRG"] = df_liestal["DELTA_AN_BRG"].dt.total_seconds()

df_liestal
```

#### 3.1.3 Verteilungsfunktionen bestimmen
Anhand der Histogramme haben wir uns eine erste Übersicht darüber verschafft, wie die Verteilungen der Zufallsvariablen etwa aussehen. Anhand von diesem Wissen, werden wir nun verschiedene Verteilungsfunktionen, welche aus unserer Sicht passen könnten, mittels der fit() Funktion von Scipy welche eine Maximum Likelihood Estimation durchführt, an unsere Daten anpassen und in einem CDF vergleichen. Anschliessend werden die Verteilungsfunktionen einzeln mit der Zufallsvariablen in einem Q-Q-Plot abgebildet. Hiermit können wir noch genauer entscheiden, welche Verteilungsfunktion sich am besten eignet.

```python
def goodness_of_fit(data, distributions):

    # Werte gleichmässig auf X-Achse verteilen
    x = np.linspace(min(data), max(data), len(data))

    # ECDF von den definierten Verteilungsfunktionen erstellen
    for distribution in distributions:
        param = eval('stats.' + distribution + '.fit(data)')
        sample = eval('stats.' + distribution + '.cdf(x, *param)')
        plt.plot(x, sample)

    # ECDF ploten
    plt.hist(data, cumulative=True, density=True, bins=200, color='#5384A6')
    plt.xlim([data.min(), data.max()])
    plt.xlabel(data.head().name)
    plt.ylabel('F(x)')
    plt.legend(distributions, loc='lower right')
    plt.title('Empirische kumulative Verteilungsfunktion von ' + data.head().name)

    # QQ-Plots für die Verteilungsfunktionen erstellen und ploten
    count = 0
    fig, ax = plt.subplots(1, len(distributions), figsize=(25, 5))
    fig.suptitle('Quantil-Quantil Plots der angepassten Verteilungsfunktionen und ' + data.head().name, fontsize=16)

    for distribution in distributions:
        param = eval('stats.' + distribution + '.fit(data)')
        sample = eval('stats.' + distribution + '.rvs(*param, size = 10000)')
        sm.qqplot_2samples(data, sample, xlabel=distribution.capitalize() + ' Verteilung', ylabel='Probe Verteilung', line='45', ax=ax[count])
        count += 1
```

##### Ankunft Olten
In den QQ- und CDF-Plot ist zu sehen, dass die Gamma und Chi-Square Verteilungsfunktionen am besten zu dieser Zufallsvariable passen. Für uns sind jedoch die höheren Werte relevant, welche zum Verpassen einer Anschlussverbindung führen. Diese Werte passen bei der <u>Exponentialverteilung</u> am besten, weswegen wir uns für diese entschieden haben.

```python
goodness_of_fit(df_liestal['DELTA_AN_OLT'], ['norm', 'expon', 'gamma', 'chi2','cauchy'])
```

##### Abfahrt Olten
Hier ist zu beobachten, dass die Chi-squared Verteilung am besten passt. Im Gegensatz zu der Zufallsvariable Ankunft Olten, sind hier die tieferen Werte wichtig, welche ebenfalls zu einem Verpassen dieser Anschlussverbindung führen. Aus diesen Gründen haben wir uns für die <u>Chi-squared Verteilung</u> entschieden.

```python
goodness_of_fit(df_liestal['DELTA_AB_OLT'], ['expon', 'norm', 'chi2', 't', 'gamma'])
```

##### Ankunft Brugg
Auch bei dieser Zufallsvariable sind die hohen Werte am wichtigsten. Da die <u>Exponentialverteilung</u> im QQ-Plot am besten passt und auch die hohen Werte relativ gut passen, haben wir uns auch hier für diese entschieden.

```python
goodness_of_fit(df_liestal['DELTA_AN_BRG'], ['expon', 't', 'cauchy'])
```

### 3.2 Basel-Brugg

#### 3.2.1 Datenaufbereitung
Da die restlichen Strecken keinen Umstieg haben, können wir diese alle gleich behandeln. Wir gehen dabei nach dem ähnlichen Prinzip wie bei der Strecke Liestal-Brugg vor, wobei wir wieder die benötigten Spalten übernehmen, die Dateitypen anpassen und anschliessend die Fahrten in einer Zeile zusammenfassen.

```python
def open_and_structure_df(path, id):

    # Daten einlesen
    df_raw = pd.read_csv(path, sep = ",")

    # Benötigte Spalten übernehmen
    df_raw = df_raw[['BETRIEBSTAG', 'HALTESTELLEN_NAME', 'LINIEN_ID','ANKUNFTSZEIT', 'AN_PROGNOSE', 'AN_PROGNOSE_STATUS']]

    # Dateitypen anpassen
    df_raw['BETRIEBSTAG'] = pd.to_datetime(df_raw['BETRIEBSTAG'], format='%d.%m.%Y')
    df_raw['ANKUNFTSZEIT'] = pd.to_datetime(df_raw['ANKUNFTSZEIT'], format='%d.%m.%Y %H:%M')
    df_raw['AN_PROGNOSE'] = pd.to_datetime(df_raw['AN_PROGNOSE'], format='%d.%m.%Y %H:%M:%S')

    #Dataframe mit allen Spalten leer erstellen
    df_structured = pd.DataFrame({
        'BETRIEBSTAG': datetime,
        'SOLL_AN_BRG': datetime,
        'IST_AN_BRG': datetime,
        'DELTA_AN_BRG': datetime}, index= [0])

    # Dataframe befüllen
    count = 0
    for i in range(0,len(df_raw)):
        summe = sum(df_raw.BETRIEBSTAG[i] == df_structured.BETRIEBSTAG)
        if summe == 0:
            df_structured.at[count, 'BETRIEBSTAG'] = df_raw.BETRIEBSTAG[i]
        if (df_raw.LINIEN_ID[i] == id) & (df_raw.HALTESTELLEN_NAME[i] == "Brugg AG"):
            ind = df_structured.index[df_structured.BETRIEBSTAG == df_raw.BETRIEBSTAG[i]][0]
            df_structured.at[ind, 'SOLL_AN_BRG'] = df_raw.ANKUNFTSZEIT[i]
            df_structured.at[ind, 'IST_AN_BRG'] = df_raw.AN_PROGNOSE[i]
        df_structured.at[i, 'DELTA_AN_BRG'] = "1"
        if summe == 0:
            count += 1
    df_structured = df_structured.dropna()
    df_structured["DELTA_AN_BRG"] = (df_structured["IST_AN_BRG"] - df_structured["SOLL_AN_BRG"])
    df_structured["DELTA_AN_BRG"] = df_structured["DELTA_AN_BRG"].dt.total_seconds()

    return df_structured
```

```python
df_basel = open_and_structure_df('./Daten/Basel-Brugg.csv', 1961)
df_basel
```

### 3.4 Umsteigezeit und Laufzeit
Da wir für die Zufallsvariablen Umsteigezeit und Laufzeit haben, wir mittels unserer Erfahrung definiert. Grundsätzlich gehen wir davon aus, dass diese Zufallsvariablen normalverteilt sind. Wir wissen, dass das Umsteigen in Olten ca. 2 Minuten dauert und der Weg vom Bahnhof Brugg bis zum LearningLab etwa 4.5 Minuten benötigt. Ausserdem können wir sagen, dass diese Werte nur leicht variieren, weshalb wir uns für eine kleine Streuung entschieden haben. Aus diesen Gründen haben wir uns für folgende Parameter entschieden:


```python
# Parameter Umsteigezeit
loc = 2 * 60    # In Sekunden umgerechnet
scale  = 6
param_umsteigezeit = loc, scale

# Parameter Laufzeit
loc = 4.5 * 60    # In Sekunden umgerechnet
scale  = 6
param_laufzeit = loc, scale
```

## 4. Wahrscheinlichkeitsmodell

```python
# Anzahl Simulationen definieren
sim_count = 1_000_000
```

### 4.1 Umsteigezeit und Laufzeit

#### 4.1.1 Monte Carlo Simulation

```python
# Zufallsvariable Umsteigezeit Olten
temp_umsteigezeit = stats.norm.rvs(*param_umsteigezeit, size = sim_count)

# Zufallsvariable Laufzeit Brugg
temp_laufzeit = stats.norm.rvs(*param_laufzeit, size = sim_count)
```

```python
# Histogramm erstellen
plt.hist(temp_laufzeit, bins=50, color='#5384A6')
plt.xlabel('Laufzeit in Sekunden')
plt.ylabel('Anzahl')
plt.title('Histrogramm von Laufzeit')
plt.show()
```

### 4.2 Strecke Liestal-Brugg

#### 4.2.1 Monte Carlo Simulation

```python
# Zufallsvariable Ankunft Olten
param = stats.expon.fit(df_liestal['DELTA_AN_OLT'])
sim_liestal_an_olt = stats.expon.rvs(*param, size = sim_count)

# Zufallsvariable Abfahrt Olten
param = stats.chi2.fit(df_liestal['DELTA_AB_OLT'])
sim_liestal_ab_olt = stats.chi2.rvs(*param, size = sim_count)

# Zufallsvariable Ankunft Brugg
param = stats.expon.fit(df_liestal['DELTA_AN_BRG'])
sim_liestal_an_brg = stats.expon.rvs(*param, size = sim_count)
```

#### 4.2.2 Simulierte Daten zusammenfügen

```python
def change_to_seconds(df,column):
    if type(df[column][0]) != int:
        for i in range(len(df)):
            if df[column][i].days < 0:
                hours_to_deduct = abs(df[column][i].days) * 86400
                df[column][i] = df[column][i].seconds - hours_to_deduct
            elif df[column][i].days >= 0:
                df[column][i] = df[column][i].seconds
    return df
```

```python
# Chained_assignment ignorieren
pd.options.mode.chained_assignment = None

# Dataframe erstellen
df_sim_liestal = pd.DataFrame({
    'SOLL_AN_OLT': datetime,
    'IST_AN_OLT': datetime,
    'DELTA_AN_OLT': datetime,
    'SOLL_AB_OLT': datetime,
    'IST_AB_OLT': datetime,
    'DELTA_AB_OLT': datetime,
    'SIM_UMSTEIGEZEIT': int,
    'DELTA_UMSTEIGEZEIT': int,
    'SOLL_AN_BRG': datetime,
    'IST_AN_BRG': datetime,
    'DELTA_AN_BRG': datetime,
    'SIM_LAUFZEIT': int,
    'UEBRIGE_ZEIT_BRUGG': int,
    'VERSPAETET_IM_UNT': pd.Series(dtype='bool')})

# Dataframe mit Zufallsvariablen befüllen
df_sim_liestal["DELTA_AN_OLT"] = sim_liestal_an_olt
df_sim_liestal["DELTA_AB_OLT"] = sim_liestal_ab_olt
df_sim_liestal["DELTA_AN_BRG"] = sim_liestal_an_brg

# Werte runden und Datentypen anpassen
df_sim_liestal["DELTA_AB_OLT"] = round(df_sim_liestal["DELTA_AB_OLT"])
df_sim_liestal["DELTA_AB_OLT"] = df_sim_liestal["DELTA_AB_OLT"].astype('int')

df_sim_liestal["DELTA_AN_OLT"] = round(df_sim_liestal["DELTA_AN_OLT"])
df_sim_liestal["DELTA_AN_OLT"] = df_sim_liestal["DELTA_AN_OLT"].astype('int')

df_sim_liestal["DELTA_AN_BRG"] = round(df_sim_liestal["DELTA_AN_BRG"])
df_sim_liestal["DELTA_AN_BRG"] = df_sim_liestal["DELTA_AN_BRG"].astype('int')

df_sim_liestal["DELTA_AN_OLT"] = df_sim_liestal["DELTA_AN_OLT"].astype('timedelta64[s]')
df_sim_liestal["DELTA_AB_OLT"] = df_sim_liestal["DELTA_AB_OLT"].astype('timedelta64[s]')
df_sim_liestal["DELTA_AN_BRG"] = df_sim_liestal["DELTA_AN_BRG"].astype('timedelta64[s]')

# Soll-Zeiten übernehmen
target_time = str(df_liestal['SOLL_AN_OLT'][0].hour) + ":" + str(df_liestal['SOLL_AN_OLT'][0].minute)
df_sim_liestal['SOLL_AN_OLT'] = pd.to_datetime(target_time, format='%H:%M')
target_time = str(df_liestal['SOLL_AB_OLT'][0].hour) + ":" + str(df_liestal['SOLL_AB_OLT'][0].minute)
df_sim_liestal['SOLL_AB_OLT'] = pd.to_datetime(target_time, format='%H:%M')
target_time = str(df_liestal['SOLL_AN_BRG'][0].hour) + ":" + str(df_liestal['SOLL_AN_BRG'][0].minute)
df_sim_liestal['SOLL_AN_BRG'] = pd.to_datetime(target_time, format='%H:%M')

# Ist-Zeiten berechnen
df_sim_liestal["IST_AB_OLT"] = (df_sim_liestal["SOLL_AB_OLT"] + df_sim_liestal["DELTA_AB_OLT"]).dt.time
df_sim_liestal["IST_AN_OLT"] = (df_sim_liestal["SOLL_AN_OLT"] + df_sim_liestal["DELTA_AN_OLT"]).dt.time
df_sim_liestal["IST_AN_BRG"] = (df_sim_liestal["SOLL_AN_BRG"] + df_sim_liestal["DELTA_AN_BRG"]).dt.time

# Datentypen anpassen
df_sim_liestal['SOLL_AB_OLT'] = df_sim_liestal['SOLL_AB_OLT'].dt.time
df_sim_liestal['SOLL_AN_OLT'] = df_sim_liestal['SOLL_AN_OLT'].dt.time
df_sim_liestal['SOLL_AN_BRG'] = df_sim_liestal['SOLL_AN_BRG'].dt.time

# Zufallsvariablen Umsteigezeit und Laufzeit einfügen
df_sim_liestal["SIM_UMSTEIGEZEIT"] = temp_umsteigezeit
df_sim_liestal["SIM_LAUFZEIT"] = temp_laufzeit

# Zu Sekunden ändern
df_sim_liestal = change_to_seconds(df_sim_liestal,"DELTA_AN_OLT")
df_sim_liestal = change_to_seconds(df_sim_liestal,"DELTA_AB_OLT")
df_sim_liestal = change_to_seconds(df_sim_liestal,"DELTA_AN_BRG")
```

#### 4.2.3 Verspätungen berechnen
Mittels dem simulierten Datensatz berechnen wir nun, bei welchen Fahrten es zu einer verspäteten Ankunft im LearningLab gekommen ist.

```python
# Zeitdifferenz zwischen Soll-Ankunftszeit und Soll-Abfahrtszeit berechnen und formatieren
minutes = (df_sim_liestal["SOLL_AB_OLT"][0].minute - df_sim_liestal["SOLL_AN_OLT"][0].minute)
hours = (df_sim_liestal["SOLL_AB_OLT"][0].hour - df_sim_liestal["SOLL_AN_OLT"][0].hour)
time = (minutes + hours * 60)*60

# Übrige Zeit berechnen
df_sim_liestal["UEBRIGE_ZEIT_OLTEN"] = time - df_sim_liestal["SIM_UMSTEIGEZEIT"] + df_sim_liestal["DELTA_AB_OLT"] - df_sim_liestal["DELTA_AN_OLT"]

# Zeitdifferenz zwischen Soll-Ankunftszeit und Unterrichtsstart berechnen und formatieren
class_start_minutes = 15 - int(df_sim_liestal["SOLL_AN_BRG"][0].minute)
class_start_hours = 8 - int(df_sim_liestal["SOLL_AN_BRG"][0].hour)
class_start_time = (class_start_minutes + class_start_hours * 60)*60

# Übrige Zeit berechnen
df_sim_liestal["UEBRIGE_ZEIT_BRUGG"] = class_start_time - df_sim_liestal["SIM_LAUFZEIT"] - df_sim_liestal["DELTA_AN_BRG"]

# Anzahl Verspätungen berechnen
late_count_liestal = 0
for i in range(len(df_sim_liestal)):
    if (df_sim_liestal["UEBRIGE_ZEIT_OLTEN"][i] < 0) | (df_sim_liestal["UEBRIGE_ZEIT_BRUGG"][i] < 0):
        late_count_liestal += 1
```

#### 4.3.2 Simulierte Daten zusammenfügen

```python
def create_sim_df(org_df,sim_data):
    # Dataframe erstellen
    df = pd.DataFrame({
        'SOLL_AN_BRG': datetime,
        'IST_AN_BRG': datetime,
        'DELTA_AN_BRG': datetime,
        'SIM_LAUFZEIT': int,
        'UEBRIGE_ZEIT_BRUGG': int,
        'VERSPAETET_IM_UNT': pd.Series(dtype='bool')})

    # Dataframe mit Zufallsvariable befüllen
    df["DELTA_AN_BRG"] = sim_data

    # Werte Runden und Datentypen anpassen
    df["DELTA_AN_BRG"] = round(df["DELTA_AN_BRG"])
    df["DELTA_AN_BRG"] = df["DELTA_AN_BRG"].astype('int')
    df["DELTA_AN_BRG"] = df["DELTA_AN_BRG"].astype('timedelta64[s]')

    # Soll-Zeit übernehmen
    target_time = str(org_df['SOLL_AN_BRG'][0].hour) + ":" + str(org_df['SOLL_AN_BRG'][0].minute)
    df['SOLL_AN_BRG'] = pd.to_datetime(target_time, format='%H:%M')

    #Ist-Zeit berechnen
    df["IST_AN_BRG"] = (df["SOLL_AN_BRG"] + df["DELTA_AN_BRG"]).dt.time

    # Datentypen anpassen
    df['SOLL_AN_BRG'] = df['SOLL_AN_BRG'].dt.time

    # Zufallsvariable Laufzeit einfügen
    df["SIM_LAUFZEIT"] = temp_laufzeit

    # Zu Sekunden ändern
    df = change_to_seconds(df,"DELTA_AN_BRG")

    return df
```

```python
df_sim_basel = create_sim_df(df_basel,sim_basel_an_brg)
```

### 4.6 Verspätungsanalyse erstellen
Um unsere zweite Fragestellung zu beantworten, berechnen wir für die von uns definierten Strecken, wie gross die Wahrscheinlichkeit für eine Verspätung im LearningLab ist, nach verschiedenen Startzeiten. Anschliessend fassen wir dies für alle Strecken in einer Tabelle zusammen.

```python
def calculate_late_arrivals(df, route):
    # Dataframe erstellen
    df_late_arrivals = pd.DataFrame({
        'Unterrichtsbeginn': range(5,16),
        'Zug': str,
        'Anzahl_Verspaetungen': int,
        'Prozente_Verspaetungen': int})

    # Durch Minuten von 08.05 - 08.15 Uhr iterieren
    for j in range(5,16):
        # Zeitdifferenz zwischen Soll-Ankunftszeit und Unterrichtsstart berechnen und formatieren
        class_start_minutes = j - int(df["SOLL_AN_BRG"][0].minute)
        class_start_hours = 8 - int(df["SOLL_AN_BRG"][0].hour)
        class_start_time = (class_start_minutes + class_start_hours * 60)*60

        # Übrige Zeit berechnen
        df["UEBRIGE_ZEIT_BRUGG"] = class_start_time - df["SIM_LAUFZEIT"] - df["DELTA_AN_BRG"]

        # Anzahl Verspätungen berechnen
        late_count = 0
        for i in range(len(df)):
            if  df["UEBRIGE_ZEIT_BRUGG"][i] < 0:
                late_count += 1

        # Dataframe befüllen
        df_late_arrivals["Anzahl_Verspaetungen"][j-5] = late_count
        df_late_arrivals["Prozente_Verspaetungen"][j-5] = late_count / sim_count * 100
        df_late_arrivals["Zug"] = route

    return df_late_arrivals
```

```python
def late_arrivals_total(df_bsl,df_olt,df_zrh):
    # Dataframe erstellen
    routes = ["","",""]
    df_late_arrivals_total = pd.DataFrame({
        'Unterrichtsbeginn (Minuten)': range(5,16),
        'Anzahl_Verspaetungen': int,
        'Verspaetungswahrscheinlichkeit (%)': int})

    # Verspätungen für die verschiedenen Strecken berechnen
    routes[0] = calculate_late_arrivals(df_bsl,"BSL-BRG")
    routes[1] = calculate_late_arrivals(df_olt,"OLT-BRG")
    routes[2] = calculate_late_arrivals(df_zrh,"ZRH-BRG")

    # Dataframe befüllen
    for i in range(11):
        sub_total = 0
        for j in range(3):
            sub_total = sub_total + routes[j]["Anzahl_Verspaetungen"][i]
        df_late_arrivals_total["Anzahl_Verspaetungen"][i] = sub_total
        df_late_arrivals_total["Verspaetungswahrscheinlichkeit (%)"][i] = round((sub_total / (sim_count * 3)) * 100,2)

    return df_late_arrivals_total
```

```python
df_total_late_arrivals = late_arrivals_total(df_sim_basel, df_sim_olten, df_sim_zuerich)
```

### 5.1 Fragestellung 1

```python
# Wahrscheinlichkeit von Verspätung berechnen
propability_liestal = round((late_count_liestal / sim_count * 100), 2)
propability_basel = round((late_count_basel / sim_count * 100), 2)

# Ausgeben
print('Die Wahrscheinlichkeiten für die beiden Strecken nach 8.15 Uhr im LearningLab anzukommen sind:')
print('Liestal-Brugg : {}%'.format(propability_liestal))
print('Basel-Brugg : {}%'.format(propability_basel))
```

### 5.2 Fragestellung 2

```python
# Barplot erstellen
fig, ax = plt.subplots()
bar_values = ax.bar(df_total_late_arrivals["Unterrichtsbeginn (Minuten)"], df_total_late_arrivals["Verspaetungswahrscheinlichkeit (%)"])
plt.axhline(y=5, color = 'red')
ax.set_ylabel('Verspätungswahrscheinlichkeit (%)')
ax.set_xlabel('Unterrichtsbeginn (Minuten)')
plt.suptitle('Verspätungswahrscheinlichkeiten nach verschiedenen Unterrichtsbeginnen', fontsize=14, fontweight='bold')
ax.set_title('Von 08:05 bis 08:15')
ax.bar_label(bar_values)
fig.tight_layout()
plt.show()
```

Wir sehen, dass der frühst mögliche Unterrichtbeginn um 8.08 Uhr wäre, wenn man beabsichtigt, dass mindestens 95% der StundentInnen püntklich im LearningLab eintreffen. Jedoch ist hinzuzufügen, dass sich die Wahrscheinlichkeit einer Verspätung, im Vergleich zu 8.15 Uhr, mindestens 30-fach erhöht. Da der Zeitgewinn nur wenige Minuten beträgt, dafür jedoch viel mehr Verspätungen entstehen, empfehlen wir nicht, die Startzeit auf 8.08 Uhr zu verschieben.
